## Description of the Original Data

These data come from a study that measured nine physiologic factors and collected 16 types of psychological data while participants made decisions in standard tasks from experimental economics measuring risk aversion, intertemporal patience, altruism, cooperation, generosity, trust and trustworthiness.   

<br>

The analysis of the original data showed that multiple measures of physiologic stress decrease cooperative behaviors while testosterone consistently increased antisocial behaviors.  Further, cognitive abilities robustly increased patience but decreased cooperation while those more satisfied with their lives were more trustworthy.  Participants scoring high on the personality trait of Imagination were shown to be less altruistic and generous, while those high on Neuroticism were more prosocial.

## Variable Explanations:

Big Five Personality Traits
* y through y4 are part of the Big Five Personality Traits, a model used in personality psychology for understanding and categorizing trait differences.
    * Extraversion relates to the degree to which a person is outgoing, social, and energized by interactions with others.
    * Agreeableness relates to the degree to which a person is kind, cooperative, and considerate.
    * Conscientiousness relates to the degree to which a person is organized, responsible, and goal-oriented.
    * Neuroticism relates to the degree to which a person experiences higher emotional instability and are more prone to negative emotions like anxiety and stress.
    * Openness relates to the degree to which a person is creative, curious, and open to new ideas and experiences.  

Subjective Well-Being
* y5 through y7 Subjective Well-Being relates to an individual's overall evaluation of their life and emotional experiences.
    * Life Satisfaction refers to an individual's judgment about the overall quality and fulfillment of their life.
    * Positive Affect involves the frequency and intensity an indivdual experiences positive emotions, such as joy, gratitude, and contentmen.
    * Negative Affect involves the frequency and intensity an indivdual experiences positive emotions, such as joy, gratitude, and contentment.
    * Negative  negative emotions, such as sadness, anxiety, and stress.


Social Economics
* y8 and y9 Dictator Game involves one participant (the dictator)who is given a sum of money and has the unilateral decision to share any portion of it with another participant (the recipient). It measures altruism (low scores), fairness (middle scores), and self-interest (high scores) by studying how dictators choose to distribute resources without external influence. Money sent reflects what the dictator chose to share.
* y10 and y11 Public Goods Game involves participants tasked with deciding how much of their resources to contribute to a shared pool to be distributed among all participants. This measures self-interest (profit score) and collective benefit (cooperation score).

Physiological Measures
* x4 Cortisol is a steroid hormone often associated with stress response.   
* x5 and x6 Estradiol and progesterone are key hormones involved in the regulation of the biologically female reproductive system. They can modulate the stress response and mood, having anxiolytic (anxiety-reducing) effects. Research suggest that levels can influence social bonding, empathy, and the recognition of social cues, and may enhance social connectedness and communication.
* x7 Testosterone is a male sex hormone that plays a key role in the development and maintenance of biologically male sexual characteristics and may influence behavior by contributing to traits such as competitiveness, aggression, and risk-taking tendencies.
