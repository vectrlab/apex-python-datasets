## Description of the Original Data


The Climate Change Hope Scale (CCHS; Li & Monroe, 2017) adapted Synder et al.'s (1991) Hope Scale to focus specifically on climate change. This scale was designed to measure willpower and waypower. Willpower is the extent to which one believes that they are able to meet their goals. Waypower is the extent to which one can think of ways to overcome a problem.

<br>

The CCHS is comprised of 11 items with each item rated on a seven-point rating from 1 (strongly disagree) to 7 (strongly agree). A prior study in an adolescent sample showed that the CCHS could be grouped into three factors: Personal-sphere willpower and waypower, Collective-sphere willpower and waypower, and Lack of willpower and waypower. This dataset was collected as part of research to test Li and Monroe's (2017) three-factor model in a sample of U.S. adults.

## Description of the Example

Access this example using the file `example.csv`.

The individual CCHS scale items and additional measures were removed to condense the dataset. The following variables are included in the example:


y: Climate Change Hope Scale (CCHS) Total Score   
y1: CCHS Personal Willpower and Waypower Subscale Score     
y2: CCHS Collective Willpower and Waypower Subscale Score    
y3: CCHS Lack of Willpower and Waypower Subscale Score          
x: Gender (1 = Female; 2 = Male; 3 = Non-Binary)    
x1: Age in years    
x2: Hispanic (1 = Yes; 2 = No)    
x3: Race (1 = Black; 2 = Asian; 3 = White; 4 = Native American; 5 = Native Hawaiian or Pacific Islander; 6 = Other)      
x4: Education in years of full time education completed    
x5: Children (1 = Has children; 2 = Does not have children)    
x6: Climate Change Doubt: Personal opinion on whether climate is changing (1 = Yes; 2 = No; 3 = Don't know)
