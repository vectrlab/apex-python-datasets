Although it may not help you with predicting current housing prices, it does provide an accessible introductory dataset for statistical analyses.


## Description of the Example

Access this example using the file `example.csv`.

The variables of longitude and latitude were removed to condense the dataset. The following variable are included in the example:


y: Median house value for households within a block (US Dollars)   
x: Median age of a house within a block; a lower number is a newer building   
x1: Total number of rooms within a block      
x2: Total number of bedrooms within a block   
x3: Total number of people residing within a block   
x4: Total number of households, a group of people residing within a home unit, for a block    
x5: Median income for households within a block of houses (tens of thousands of US Dollars)    
x6: Location of the house with respect to the ocean/sea
